```python
df['date']=pd.to_datetime(df['date'])
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
```

```python
df.columns
```

```python
df.drop(['id','date','lat','long'],axis=1,inplace=True)
```

```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=11)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier=Sequential()
classifier.add(Dense(19,activation='relu'))
classifier.add(Dense(5,activation='relu'))
classifier.add(Dense(1))
classifier.compile(optimizer='adam',loss='mse')
classifier.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=400)
```

```python
loss_df=pd.DataFrame(classifier.history.history)
loss_df.plot()
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
pred=classifier.predict(x_test)
```

```python
mean_absolute_error(y_test,pred)
```

```python
r2_score(y_test,pred)
```

```python
plt.plot(y_test,y_test,'r')
plt.scatter(y_test,pred)
```
